this code is combining dataset download the drug <-> with LUAD and LUSC (seperate database)

```python
import pandas as pd 
import numpy as np 
import selfies as sf 
import pubchempy as pcp 
```

```python
LUAD_path = 'dataset/LUAD_drug_to_features.csv'
LUSC_path = 'dataset/LUSC_drug_to_features.csv'
dfLUAD = pd.read_csv(LUAD_path)
dfLUSC = pd.read_csv(LUSC_path)

print(dfLUAD.head(10))

print("columns: ", dfLUAD.columns)
```

```python
#count the number of unique drugs in LUAD and LUSC
print("LUAD unique drugs: ", dfLUAD['Drug name'].nunique())
print("LUSC unique drugs: ", dfLUSC['Drug name'].nunique())

#make list of unique drugs in LUAD and LUSC
LUAD_drugs = dfLUAD['Drug name'].unique()
LUSC_drugs = dfLUSC['Drug name'].unique()

print("LUAD drugs: ", LUAD_drugs)
```

```python
dfSMILE = []
for drug in LUAD_drugs:
    try:
        compound = pcp.get_compounds(drug, 'name')[0]
        smiles = compound.canonical_smiles
        selfies = sf.encoder(smiles)
        dfSMILE.append([drug, smiles, selfies])
    except:
        print("Error: ", drug)
        continue
    
    
print(dfSMILE)
```

```python
print(dfSMILE)
dfSMILE = pd.DataFrame(dfSMILE, columns=['Drug name', 'SMILES', 'SELFIES'])
dfSMILE.to_csv('dataset/LUAD_drug_to_SMILES.csv', index=False)
```

```python
new_data = pd.read_csv('dataset/LUAD_drug_to_SMILES.csv')
new_data
```

```python
# Merge the new data with the original data
merged_data = pd.merge(dfLUAD, new_data, on='Drug name', how='left')
# Save the merged data to a new CSV file
merged_data.to_csv('dataset/LUAD_drug_to_features_SMILES.csv', index=False)
# Print the merged data
#print(merged_data.head(10))
# drop columns that are not needed and make it left only Drug Name, SMILES, SELFIES, and Drug ID,ic50_effect_size 
merged_data = merged_data[['Drug name', 'Feature Name', 'SMILES', 'SELFIES', 'Drug ID', 'ic50_effect_size','Tissue Type']]
print(merged_data.head(10))
```

```python
#save the merged data to a new CSV file *********
merged_data.to_csv('dataset/LUAD_drug_SMILES_SELFIES.csv', index=False)
```

```python
# Do the same for LUSC
dfLUSC_SMILE = []
for drug in LUSC_drugs:
    try:
        compound = pcp.get_compounds(drug, 'name')[0]
        smiles = compound.canonical_smiles
        selfies = sf.encoder(smiles)
        dfLUSC_SMILE.append([drug, smiles, selfies])
    except:
        print("Error: ", drug)
        continue

#print(dfLUSC_SMILE)
dfLUSC_SMILE = pd.DataFrame(dfLUSC_SMILE, columns=['Drug name', 'SMILES', 'SELFIES'])
```

```python
print(dfLUSC_SMILE)
```

```python
#merge the new data with the original data
merged_data_LUSC = pd.merge(dfLUSC, dfLUSC_SMILE, on='Drug name', how='left')
# Save the merged data to a new CSV file
merged_data_LUSC.to_csv('dataset/LUSC_drug_SMILES_SELFIES.csv', index=False)
#drop columns that are not needed and make it left only Drug Name, SMILES, SELFIES, and Drug ID,ic50_effect_size
merged_data_LUSC = merged_data_LUSC[['Drug name', 'Feature Name', 'SMILES', 'SELFIES', 'Drug ID', 'ic50_effect_size','Tissue Type']]
print(merged_data_LUSC.head(10))
```

```python
#display dataframe with 10 rows of LUSC_drug_SMILES_SELFIES.csv
print(merged_data_LUSC.head(10))
```

```python
#merge data between LUSC_drug_SMILES_SELFIES.csv and LUAD_drug_SMILES_SELFIES.csv
merged_data_all = pd.concat([merged_data, merged_data_LUSC], ignore_index=True)
print(merged_data_all.head(10))

#drop columns SMILE and SELFIES that have all NaN values
merged_data_all = merged_data_all.dropna(subset=['SMILES', 'SELFIES'], how='all')
#save the merged data to a new CSV file
merged_data_all.to_csv('dataset/LUAD_LUSC_drug_finalPreprocess.csv', index=False)
```
