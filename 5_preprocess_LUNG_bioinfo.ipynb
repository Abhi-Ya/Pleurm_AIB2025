This code notebook aim to prep the gene expression, gene mutation file to be ready for the GNN training

```python
#import path of file from original_GeneExpression and original_GeneMutation
expression_path = '5_dataset/original_GeneExpression.csv'
mutation_path = '5_dataset/original_GeneMutation.csv'
import pandas as pd
import numpy as np

# Load the data
expression_data = pd.read_csv(expression_path, index_col=0)
mutation_data = pd.read_csv(mutation_path, index_col=0)
```

```python
expression_data.head(3)
# y-axis is CCL name list, x-axis is gene name list
```

-------------------GENE EXPRESSION PROCESSING----------------------------------------

```python
#delete the column chr, fpos, tpos, strand
mutation_data = mutation_data.drop(['chr','fpos','tpos','strand'], axis=1)
```

```python
#switch row and column of mutation data
mutation_data = mutation_data.T
```

```python
mutation_data.head(5)
```

```python
#find if the row name have the word "LUNG" in the row name, if yes then stay, if no then delete the row
lung_mutation_data = mutation_data[mutation_data.index.str.contains('LUNG')]
lung_mutation_data.head(5)
#report the number of rows and columns of lung_mutation_data
lung_mutation_data.shape
```

```python
#remove the "_LUNG" of the row name
lung_mutation_data.index = lung_mutation_data.index.str.replace('_LUNG', '', regex=False)
#remove the detail after the first "_" of column name 
lung_mutation_data.columns = lung_mutation_data.columns.str.split('_').str[0]
#report the number of rows and columns of lung_mutation_data
lung_mutation_data.shape
```

```python
lung_mutation_data.head(5)
#y-axis is gene name list, x-axis is CCL name list
```

```python
#save lung_mutation_data
lung_mutation_data.to_csv('5_dataset/lung_mutation_data_1.csv', index=True)
```

```python
CCL_list_path = 'dataset/cell lines in Lung.csv'
CCL_lung_list = pd.read_csv(CCL_list_path)
CCL_lung_list.head()
```

```python
# reload because my computer died.
lung_mutation_data = pd.read_csv('5_dataset/lung_mutation_data_1.csv')
#rename first column to "cell line name"
lung_mutation_data.rename(columns={'Unnamed: 0': 'cell line name'}, inplace=True)
lung_mutation_data.head(5)
```

```python
# append the 'Depmap Id' column from CCL_lung_list to lung_mutation_data if the 'cell line name' in lung_mutation_data is the same as 'Cell Line' column in CCL_lung_list
lung_mutation_data_ID = lung_mutation_data.merge(CCL_lung_list[['Cell Line', 'Depmap Id']], left_on='cell line name', right_on='Cell Line', how='left')
lung_mutation_data_ID
```

```python
#move Depmap Id to the second column
lung_mutation_data_ID = lung_mutation_data_ID[['cell line name', 'Depmap Id'] + [col for col in lung_mutation_data_ID.columns if col not in ['cell line name', 'Depmap Id']]]
#delete the column Cell Line
lung_mutation_data_ID = lung_mutation_data_ID.drop(['Cell Line'], axis=1)
```

```python
#save lung_mutation_data_ID
lung_mutation_data_ID.to_csv('5_dataset/lung_mutation_data_ID.csv')
lung_mutation_data_ID.head(5)

```

---------------------------GENE EXPRESSION PREPROCESSING-----------------------------

AIM: make the row corresponse to same data as gene mutation data (LUNG)

```python
#read the column "CCL/Gene" in expression_data
expression_data = pd.read_csv(expression_path)
```

```python
expression_data.head(5)
```

```python
#delete the row in expression_data if the the column 'CCL/Gene' is not in the column 'cell line name' of lung_mutation_data_ID
expression_data = expression_data[expression_data['CCL/Gene'].isin(lung_mutation_data_ID['Depmap Id'])]
expression_data.head(5)
#report the size of expression_data
expression_data.shape
```

```python
#delete row in lung_mutation_data_ID if the column 'Depmap Id' is not in the column 'CCL/Gene' of expression_data
lung_mutation_data_ID = lung_mutation_data_ID[lung_mutation_data_ID['Depmap Id'].isin(expression_data['CCL/Gene'])]
#report the size of lung_mutation_data_ID
lung_mutation_data_ID.shape
```

```python
#save expression_data
expression_data.to_csv('5_dataset/final_lung_expression_data.csv', index=True)
#save lung_mutation_data_ID
lung_mutation_data_ID.to_csv('5_dataset/final_lung_mutation_data.csv', index=True)
```
